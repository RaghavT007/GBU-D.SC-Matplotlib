## Simple Line Plot

```python
# Line Plot showing active, confirmed, recovered & deceased cases
```

```python
A = data.iloc[0:,1].values
R = data.iloc[0:,4].values
D = data.iloc[0:,3].values
C = data.iloc[0:,2].values
X = data.iloc[0:,0]

#For designating the data values in a alphabet.
```

```python
plt.plot(X,A, label="Active cases", color="red", linewidth=2, marker='+')
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('COVID-19 cases')
plt.legend()
plt.show()

#Line plot between District(X) and No.of cases(Y) representing Active cases
```

```python
plt.plot(X,C, label="Confirmed cases", color="turquoise", linewidth=2, marker='o')
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('COVID-19 cases')
plt.legend()
plt.show()

#Line plot between District(X) and No.of cases(Y) representing Confirmed cases
```

```python
plt.plot(X,R, label="Recovered cases", color="green", linewidth=2, marker='*')
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('COVID-19 cases')
plt.legend()
plt.show()

#Line plot between District(X) and No.of cases(Y) representing Recovered cases
```

```python
plt.plot(X,D, label="Deceased cases", color="black", linewidth=2, marker='x')
plt.ylabel('No. of cases')
plt.title('COVID-19 cases')
plt.legend()
plt.show()

#Line plot between District(X) and No.of cases(Y) representing Deceased cases
```

```python
plt.plot(X,A, label="Active cases", color="red", linewidth=2, marker='+')
plt.plot(X,C, label="Confirmed cases", color="turquoise", linewidth=2, marker='o')
plt.plot(X,R, label="Recovered cases", color="green", linewidth=2, marker='*')
plt.plot(X,D, label="Deceased cases", color="black", linewidth=2, marker='x')
plt.xlabel('District')
plt.ylabel('No. of cases')
plt.title('COVID-19 cases')
plt.legend()
plt.show()

#Single line plot between District(X) and No.of cases(Y) representing all active, confirmed, recovered & deceased cases
```

```python
#Plot active vs recovered cases for top 5 districts having highest number of cases
```

```python
data.sort_values(['Active','District'], ascending = False)

#Sort all the Active data with district in descending order
```

```python
sortcases= data.sort_values(['Active','District'], ascending = False)
sortcases.head(5)

#Sort 5 Active data with district in descending order
```

```python
highestcases = sortcases.head(5)

a = highestcases.loc[:,"Active"]
r = highestcases.loc[:,"Recovered"]
plt.bar(a,r, width = 150, color='yellow')
plt.xlabel("Active cases")
plt.ylabel("Recovered cases")
plt.title("Highest Covid-19 cases in top 5 districts")
plt.show()

#Bar plot between Active vs Recovered cases of top 5 highest active COVID-19 cases
```

```python
plt.scatter(a,r)
plt.xlabel("Active cases")
plt.ylabel("Recovered cases")
plt.title("Scattered plot of Active vs Recovered cases of Top 5 districts")
plt.tight_layout()
plt.show()

#Scatter plot between Active vs Recovered cases of top 5 highest active COVID-19 cases district
```

```python
labels= ['Raipur', 'Durg', 'Rajnandgaon', 'Bilaspur', 'Bastar']
colors = ['red', 'crimson', 'palevioletred', 'lightpink', 'mistyrose']
plt.pie(a, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Active cases in Top 5 Districts')
plt.show()

#Pie chart of top 5 highest active COVID-19 cases district
```
